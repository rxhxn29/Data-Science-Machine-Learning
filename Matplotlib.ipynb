```python
import numpy as np
```

```python
x= np.linspace(0,10,100)
```

```python
y= np.sin(x)
z=np.cos(x)
print(x,y,z)
```

# Plot the data

```python
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
#adding title to x and y axis
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('y and z')
plt.title("test")
plt.show()
```

```python
#ploting Parabola
x =np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,"r+")
plt.show()
```

## Bar plots

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.bar(langs,students)
plt.show()
```

## Pie Chart

```python
fig2 = plt.figure()
ax2 = fig2.add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax2.pie(students,labels=langs,autopct='%1.2f%%')
```

## Scatter Plot

```python
x = np.linspace(0,10,100)
y= np.sin(x)
z=np.cos(x)
plt.scatter(x,y)
plt.show()
```

## 3D scatter graph

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
z=np.linspace(0,15,1000)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Greens')
plt.show()
```
